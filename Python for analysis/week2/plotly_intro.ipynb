```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy.random import exponential
%matplotlib inline
```

```python
fig, ax = plt.subplots()
ax.plot(exponential(5, 50))
ax.set_title('Graphic')
ax.set_xlim(20)
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 3, figsize = (10, 5))
for row, row_axes in enumerate(ax):
    for column, ax in enumerate(row_axes):
        ax.plot(exponential(column, 30))
        ax.set_title(' {}  {}  '.format(column+1, row+1))
fig.tight_layout()
```

```python
fig.savefig('results.png')
```
