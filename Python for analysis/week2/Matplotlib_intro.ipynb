```python
import pandas as pd
from numpy.random import exponential
```

```python
df = pd.DataFrame({'x':range(20), 'y':exponential(10,20)})
```

```python
%matplotlib inline
```

```python
df.y.hist()
```

```python
import matplotlib.pyplot as plt
```

```python
data1 = exponential(5, 20)
data2 = exponential(6, 20)
```

```python
plt.plot(data1,  label = '1st')
plt.scatter(range(20), data2, color = 'red', label = '2nd')
plt.title('Результаты', fontdict = {'fontsize':20})
plt.xlabel('X')
plt.ylabel('Результат')
plt.legend()
plt.xticks(range(0, 20, 5))
```

```python
plt.savefig('results.png')
```
