# Непрерывные распределения

```python
a = 10
b = 15
```

```python
rv_uniform = stats.uniform(a, b-a)
```

```python
rv_uniform.cdf(12) # значение функций распределения
```

```python
rv_uniform.pdf(12) # значение функций плотности вероятности
```

```python
X = np.linspace(a-2, b+2, 1000)
cdf = rv_uniform.cdf(X)
plt.plot(X, cdf)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim([-0.2, 1.2])
plt.title('Distribution function for uniform')
```

```python
X = np.linspace(a-2, b+2, 1000)
pdf = rv_uniform.pdf(X)
plt.plot(X, pdf)
plt.xlabel('x')
plt.ylabel('f(x)')
#plt.ylim([-0.2, 1.2])
plt.title('PDF for uniform')
```

# Нормальное распределение

```python
mu = 2
sigma = 1
```

```python
rv_norm = stats.norm(loc = mu, scale = sigma)
```

```python
rv_norm.rvs(15)
```

```python
X = np.linspace(0,5, 1000)
cdf = rv_norm.cdf(X)
plt.plot(X, cdf)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim([0, 1])
plt.title('Distribution function for norm')
```

```python
X = np.linspace(0,5, 1000)
pdf = rv_norm.pdf(X)
plt.plot(X, pdf)
plt.xlabel('x')
plt.ylabel('f(x)')
#plt.ylim([0, 1])
plt.title('PDF for norm')
```

```python
mu = 10
sigmas = [1, 2, 4, 6]
X = np.linspace(0, 20, 100)
for sigma in sigmas:
    rv_norm = stats.norm(loc = mu, scale = sigma)
    cdf = rv_norm.cdf(X)
    plt.plot(X, cdf, label = sigma)
plt.legend()
```

```python
mu = 10
sigmas = [1, 2, 4, 6]
X = np.linspace(0, 20, 100)
for sigma in sigmas:
    rv_norm = stats.norm(loc = mu, scale = sigma)
    pdf = rv_norm.pdf(X)
    plt.plot(X, pdf, label = sigma)
plt.legend()
```
